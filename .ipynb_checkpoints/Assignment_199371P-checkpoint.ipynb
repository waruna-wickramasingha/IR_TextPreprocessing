# 1.Tokenizing each of the files

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

#### 1.1 Tokenizing twitter feed

```python
file = open("twitter_feed.txt", mode='r', encoding="Latin-1")
twitterFeed = file.read()
file.close()

twitterTokens = word_tokenize(twitterFeed)

twitterTokensOutFile = open("Tokens/ExtractedTokens/Tokens_twitter_feed.txt", mode='w+', encoding="Latin-1")

for token in twitterTokens:
    twitterTokensOutFile.write(token + "\n")
    
twitterTokensOutFile.close()
```

#### 1.2 Tokenizing student course feedback

```python
file = open("student_cource_feedback.txt", mode='r', encoding="utf-8")
studentFeedback = file.read()
file.close()

studentFeedbackTokens = word_tokenize(studentFeedback)

studentFeedbackTokensOutFile = open("Tokens/ExtractedTokens/Tokens_studentFeedback.txt", mode='w+', encoding="utf-8")

for token in studentFeedbackTokens:
    studentFeedbackTokensOutFile.write(token + "\n")
    
studentFeedbackTokensOutFile.close()
```

#### 1.3 Tokenizing research papaer

```python
file = open("reserach_paper.txt", mode='r')
researchPaper = file.read()
file.close()

researchPaperTokens = word_tokenize(researchPaper)

researchPaperTokensOutFile = open("Tokens/ExtractedTokens/Tokens_researchPaper.txt", mode='w+')

for token in researchPaperTokens:
    researchPaperTokensOutFile.write(token + "\n")
    
researchPaperTokensOutFile.close()
```

# 2. Isolated word correction on each of the text

#### 2.1 Twitter tokens isolated correction

```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker(distance=2)
twitterMisspelledTokens = spell.unknown(twitterTokens)

print("Twitter misspelled tokens count = ", len(twitterMisspelledTokens))

twitterMisspelledFile = open('Tokens/IsolatelyCorrectedTokens/TwitterIsolatelyCorrectedTokens.csv', mode='w+', encoding="Latin-1")
twitterMisspelledFile.write("Missplled Token,Corrected Token\n")

for token in twitterMisspelledTokens:
    twitterMisspelledFile.write(token + ",")
    twitterMisspelledFile.write(spell.correction(token) + "\n")

twitterMisspelledFile.close()
```

#### 2.2 Student course feedback tokens isolated correction

```python
studentFeedbackMisspelledTokens = spell.unknown(studentFeedbackTokens)

print("Student feedback misspelled tokens count = ", len(studentFeedbackMisspelledTokens))

studentMisspelledFile = open('Tokens/IsolatelyCorrectedTokens/StudentFeedbackIsolatelyCorrectedTokens.csv', mode='w+', encoding="utf-8")
studentMisspelledFile.write("Missplled Token,Corrected Token\n")

for token in studentFeedbackMisspelledTokens:
    studentMisspelledFile.write(token + ",")
    studentMisspelledFile.write(spell.correction(token) + "\n")

studentMisspelledFile.close()
```

#### 2.3 Research Paper tokens isolated correction

```python
researchPaperMisspelledTokens = spell.unknown(researchPaperTokens)

print("Research Paper misspelled tokens count = ", len(researchPaperMisspelledTokens))

researhPaperMisspelledFile = open('Tokens/IsolatelyCorrectedTokens/ResearchPaperIsolatelyCorrectedTokens.csv', mode='w+')
researhPaperMisspelledFile.write("Missplled Token,Corrected Token\n")

for token in researchPaperMisspelledTokens:
    researhPaperMisspelledFile.write(token + ",")
    researhPaperMisspelledFile.write(spell.correction(token) + "\n")

researhPaperMisspelledFile.close()
```

## 3. Context sensitive word correction

```python
from symspellpy.symspellpy import SymSpell, Verbosity
```

#### 3.1 Creating a dictionary from the corpus "big.txt"

"big.txt" was downloaded from "How to Write a Spelling Corrector" by Peter Norvig from  http://norvig.com/spell-correct.html. It is a concatenation of public domain book excerpts from Project Gutenberg and lists of most frequent words from Wiktionary and the British National Corpus.


```python
max_edit_distance_dictionary = 2
prefix_length = 7

sym_spell = SymSpell(max_edit_distance_dictionary, prefix_length)

# create dictionary using big.txt
if not sym_spell.create_dictionary("Dictionary/big.txt"):
    print("Corpus file not found")
```

#### 3.2 Context sensitive word correction for Twitter feed

```python
twitterCntxtSensitiveCorrectionFile = open('Tokens/ContextSensitiveCorrections/TwitterContextSensitiveCorrections.csv', mode='w+', encoding="Latin-1")
twitterCntxtSensitiveCorrectionFile.write("Incorrect Biword," + "Corrected Result," + "Edit Distance Sum" + "\n")

for index in range(len(twitterTokens) - 1):
    biword = twitterTokens[index] + " "  + twitterTokens[index+1]
    result = sym_spell.word_segmentation(biword)
    
    if result.corrected_string.lower() != biword.lower():
        twitterCntxtSensitiveCorrectionFile.write(biword + "," + result.corrected_string + "," + str(result.distance_sum) + "\n")
    else:
        print("Correct Biwords >>", biword)
        
twitterCntxtSensitiveCorrectionFile.close()
```

#### 3.3 Student Feedback data context sensitive correction

```python
studentCntxtSensitiveCorrectionFile = open('Tokens/ContextSensitiveCorrections/StudentFeedbackContextSensitiveCorrections.csv', mode='w+', encoding="utf-8")
studentCntxtSensitiveCorrectionFile.write("Incorrect Biword," + "Corrected Result," + "Edit Distance Sum" + "\n")

for index in range(len(studentFeedbackTokens) - 1):
    biword = studentFeedbackTokens[index] + " "  + studentFeedbackTokens[index+1]
    result = sym_spell.word_segmentation(biword)
    
    if result.corrected_string.lower() != biword.lower():
        studentCntxtSensitiveCorrectionFile.write(biword + "," + result.corrected_string + "," + str(result.distance_sum) + "\n")
    else:
        print("Correct Biwords >>", biword)
        
studentCntxtSensitiveCorrectionFile.close()
```

#### 3.4 Research paper context sensitive correction

```python
ReseachPaperCntxtSensCorrectFile = open('Tokens/ContextSensitiveCorrections/ReseachPaperContextSensitiveCorrections.csv', mode='w+', encoding="utf-8")
ReseachPaperCntxtSensCorrectFile.write("Incorrect Biword," + "Corrected Result," + "Edit Distance Sum" + "\n")

for index in range(len(researchPaperTokens) - 1):
    biword = researchPaperTokens[index] + " "  + researchPaperTokens[index+1]
    result = sym_spell.word_segmentation(biword)
    
    if result.corrected_string.lower() != biword.lower():
        ReseachPaperCntxtSensCorrectFile.write(biword + "," + result.corrected_string + "," + str(result.distance_sum) + "\n")
    else:
        print("Correct Biwords >>", biword)
        
ReseachPaperCntxtSensCorrectFile.close()
```

# 4. Stemming using Porter's stemmer

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

#### 4.1 Twitter feed data stemming

```python
twitterFeedStemmed = [stemmer.stem(token) for token in twitterTokens]

twitterFeedStemmedFile = open("Tokens/StemmedResults/TwitterFeedStemmedResults.csv", mode="w+", encoding="Latin-1")
twitterFeedStemmedFile.write("Original Token," + "Stemmed Token\n")

for index in range(len(twitterFeedStemmed)):
    twitterFeedStemmedFile.write(twitterTokens[index] + "," + twitterFeedStemmed[index] + "\n")
    
twitterFeedStemmedFile.close()
```

#### 4.2 Student Feedback Data stemming

```python
studentFeedbackStemmed = [stemmer.stem(token) for token in studentFeedbackTokens]

studentFeedbackStemmedFile = open("Tokens/StemmedResults/StudentFeedbackStemmedResults.csv", mode="w+", encoding="utf-8")
studentFeedbackStemmedFile.write("Original Token," + "Stemmed Token\n")

for index in range(len(studentFeedbackStemmed)):
    studentFeedbackStemmedFile.write(studentFeedbackTokens[index] + "," + studentFeedbackStemmed[index] + "\n")
    
studentFeedbackStemmedFile.close()
```

#### 4.3 Research Paper data stemming

```python
reseachPaperStemmed = [stemmer.stem(token) for token in researchPaperTokens]

reseachPaperStemmedFile = open("Tokens/StemmedResults/reseachPaperStemmedResults.csv", mode="w+", encoding="utf-8")
reseachPaperStemmedFile.write("Original Token," + "Stemmed Token\n")

for index in range(len(reseachPaperStemmed)):
    reseachPaperStemmedFile.write(researchPaperTokens[index] + "," + reseachPaperStemmed[index] + "\n")
    
reseachPaperStemmedFile.close()
```

# 5. Lemmatization 
NOTE: WordNet Lemmatizer is used for this purpose

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

#### 5.1 Lemmatizing twitter feed data

```python
twitterFeedLemmatized = [wordnet_lemmatizer.lemmatize(word) for word in twitterTokens]

twitterFeedLemmatizedFile = open("Tokens/LemmatizedResults/TwitterFeedLemmatizedResults.csv", mode="w+", encoding="Latin-1")
twitterFeedLemmatizedFile.write("Original Token," + "Lemmatized Token\n")

for index in range(len(twitterFeedLemmatized)):
    twitterFeedLemmatizedFile.write(twitterTokens[index] + "," + twitterFeedLemmatized[index] + "\n")
    
twitterFeedLemmatizedFile.close()
```

#### 5.2  Lemmatizing student feedback data

```python
studentFeedbackLemmatized = [wordnet_lemmatizer.lemmatize(word) for word in studentFeedbackTokens]

studentFeedbackLemmatizedFile = open("Tokens/LemmatizedResults/StudentFeedbackLemmatizedResults.csv", mode="w+", encoding="utf-8")
studentFeedbackLemmatizedFile.write("Original Token," + "Lemmatized Token\n")

for index in range(len(studentFeedbackLemmatized)):
    studentFeedbackLemmatizedFile.write(studentFeedbackTokens[index] + "," + studentFeedbackLemmatized[index] + "\n")
    
studentFeedbackLemmatizedFile.close()
```

#### 5.3 Lemmatizing Research paper data

```python
researchPaperLemmatized = [wordnet_lemmatizer.lemmatize(word) for word in researchPaperTokens]

researchPaperLemmatizedFile = open("Tokens/LemmatizedResults/ResearchPaperLemmatizedResults.csv", mode="w+", encoding="utf-8")
researchPaperLemmatizedFile.write("Original Token," + "Lemmatized Token\n")

for index in range(len(researchPaperLemmatized)):
    researchPaperLemmatizedFile.write(researchPaperTokens[index] + "," + researchPaperLemmatized[index] + "\n")
    
researchPaperLemmatizedFile.close()
```
